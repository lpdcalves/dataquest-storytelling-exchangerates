```python
import pandas as pd
try:
    exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
except:
    print("O arquivo {} não existe".format('euro-daily-hist_1999_2020.csv'))
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

### Informações básicas do dataset
Esse dataset trata da taxa de cambio de várias moedas em relação ao dolar.

Como pode-se ver, o número de colunas do dataset é 41, com cada coluna sendo uma moeda diferente de um país diferente. Com um total de 5699 entradas no dataset, pode-se notar também que algumas moedas tem valores nulos, como é o caso inclusive do Real brasileiro que só tem 5431 valores não nulos dos 5699 possíveis. Por fim, outra informação notável é que a maioria dos dados foi armazenado em formato de string, enquanto que apenas três moedas foram armazenadas em float.

### Data Cleaning

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

### Visualizando os dados através de gráficos

Abaixo geramos um gráfico da evolução temporal da taxa de variação do euro em relação ao dolar. No entanto, há muita variação no gráfico, devido a variação dia-a-dia da cotação da moeda.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

Como alternativa a um gráfico com muita variação diária, podemos utilizar uma média móvel. Abaixo implemento uma média móvel de 7, 30, 100 e 365 dias.

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'].rolling(7).mean())
plt.title('Média móvel 7 dias', weight='bold')
plt.show()
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'].rolling(30).mean())
plt.title('Média móvel 30 dias', weight='bold')
plt.show()
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'].rolling(100).mean())
plt.title('Média móvel 100 dias', weight='bold')
plt.show()
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'].rolling(365).mean())
plt.title('Média móvel 365 dias', weight='bold')
plt.show()
```

Por fim, vamos pegar a média móvel de 30 dias e armazenar em outra coluna de `euro_to_dollar`

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

## Storytelling a partir de gráficos

### Seguindo o projeto guiado dataquest

Nessa primeira parte vamos seguir o projeto guiado do dataquest que utiliza os dados de cotação do euro e dolar para gerar um gráfico de linha que represente a variação dessa cotação para cada presidente americano desde George W. Bush.

```python
bush_obama_trump = euro_to_dollar.copy(
                    )[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy(
       )[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

### Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Adding the subplots
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

### Changes to all the subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],
                   alpha=0.3)
    ax.grid(alpha=0.5)    
    

### Ax1: Bush
ax1.plot(bush['Time'], bush['rolling_mean'],
        color='#BF5FFF')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'],
                   alpha=0.3)
ax1.text(731516.0, 1.92, 'BUSH', fontsize=18, weight='bold',
        color='#BF5FFF')
ax1.text(731216.0, 1.8, '(2001-2009)', weight='bold',
        alpha=0.3)


### Ax2: Obama
ax2.plot(obama['Time'], obama['rolling_mean'],
        color='#ffa500')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015', '', '2017'],
                   alpha=0.3)
ax2.text(734288.0, 1.92, 'OBAMA', fontsize=18, weight='bold',
        color='#ffa500')
ax2.text(734138.0, 1.8, '(2009-2017)', weight='bold',
         alpha=0.3)


### Ax3: Trump
ax3.plot(trump['Time'], trump['rolling_mean'],
        color='#00B2EE')
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '',
                     '2020', '', '2021'],
                   alpha=0.3)
ax3.text(736855.0, 1.92, 'TRUMP', fontsize=18, weight='bold',
        color='#00B2EE')
ax3.text(736745.0, 1.8, '(2017-2021)', weight='bold',
         alpha=0.3)

### Ax4: Bush-Obama-Trump
ax4.plot(bush['Time'], bush['rolling_mean'],
        color='#BF5FFF')
ax4.plot(obama['Time'], obama['rolling_mean'],
        color='#ffa500')
ax4.plot(trump['Time'], trump['rolling_mean'],
        color='#00B2EE')
ax4.grid(alpha=0.5)
ax4.set_xticks([])

### Adding a title and a subtitle
ax1.text(730016.0, 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(730016.0, 2.14, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''',
        fontsize=16)

### Adding a signature
ax4.text(729916.0, 0.65, '©DATAQUEST' + ' '*103 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

plt.show()
```

## Criando gráficos para o Real Brasileiro

Vamos analisar o dataset original e limpar os dados relacionados à cotação do real brasileiro para poder gerar um gráfico com a progressão temporal da cotação ao longo do mandato de diferentes presidentes brasileiros.

```python
exchange_rates.info()
```

Mudando o nome da coluna para BRL_real para facilitar a programação

```python
exchange_rates.rename(columns={'[Brazilian real ]': 'BRL_real'}, inplace=True)
```

```python
exchange_rates.head()
```

```python
exchange_rates['BRL_real'].value_counts()
```

Similarmente ao dolar, a tabela de real possui muitas entradas inválidas. Vamos removê-las:

```python
euro_to_dollar_to_real = euro_to_dollar.copy()
euro_to_dollar_to_real.head()
```

```python
euro_to_dollar_to_real['BRL_real'] = exchange_rates['BRL_real']
```

```python
euro_to_dollar_to_real = euro_to_dollar_to_real[euro_to_dollar_to_real['BRL_real'] != '-']
```

```python
euro_to_dollar_to_real['BRL_real'] = euro_to_dollar_to_real['BRL_real'].astype(float)
```

```python
euro_to_dollar_to_real.info()
```

Criamos um novo vetor com a cotação do dolar e real em relação ao euro chamado `euro_to_dollar_to_real` e limpamos a coluna do real.

```python
plt.plot(euro_to_dollar_to_real['Time'],
         euro_to_dollar_to_real['BRL_real'].rolling(100).mean())
plt.title('Real Brasileiro em Euros', weight='bold')
plt.show()
```

Criando um novo dataset para conter somente a taxa de cambio do euro e do dollar, agora em relação ao Real brasileiro.

```python
brl_to_euro_to_dollar = euro_to_dollar_to_real[['Time', 'BRL_real']].copy()
brl_to_euro_to_dollar.rename(columns={'BRL_real': 'euro_rate'}, inplace=True)
brl_to_euro_to_dollar['dollar_rate'] = euro_to_dollar_to_real['BRL_real']/euro_to_dollar_to_real['US_dollar']
```

```python
brl_to_euro_to_dollar.info()
```

```python
plt.plot(brl_to_euro_to_dollar['Time'],
         brl_to_euro_to_dollar['euro_rate'].rolling(100).mean())
plt.title('Real Brasileiro em Euros', weight='bold')
plt.show()
```

```python
plt.plot(brl_to_euro_to_dollar['Time'],
         brl_to_euro_to_dollar['dollar_rate'].rolling(100).mean())
plt.title('Real Brasileiro em Dolares', weight='bold')
plt.show()
```

### Gerando gráficos para a cotação durante os mandatos presidenciais

```python
brl_to_euro_to_dollar['dollar_rolling_mean'] = brl_to_euro_to_dollar['dollar_rate'].rolling(50).mean()
brl_to_euro_to_dollar['euro_rolling_mean'] = brl_to_euro_to_dollar['euro_rate'].rolling(50).mean()
```

```python
todos_presidentes = brl_to_euro_to_dollar.copy(
                   )[(brl_to_euro_to_dollar['Time'].dt.year >= 2000) & (brl_to_euro_to_dollar['Time'].dt.year < 2021)]
fhc = todos_presidentes.copy(
       )[todos_presidentes['Time'].dt.year < 2002]
lula = todos_presidentes.copy(
       )[(todos_presidentes['Time'].dt.year >= 2002) & (todos_presidentes['Time'].dt.year < 2010)]
dilma = todos_presidentes.copy(
       )[(todos_presidentes['Time'].dt.year >= 2010) & ((todos_presidentes['Time'].dt.year < 2017) & (todos_presidentes['Time'].dt.month < 9))]
temer = todos_presidentes.copy(
       )[((todos_presidentes['Time'].dt.year >= 2016) & (todos_presidentes['Time'].dt.month >= 9))  & (todos_presidentes['Time'].dt.year < 2019)]
bozo = todos_presidentes.copy(
       )[(todos_presidentes['Time'].dt.year >= 2019)  & (todos_presidentes['Time'].dt.year < 2022)]
```

```python
plt.plot(lula['Time'],
         lula['dollar_rolling_mean'])
plt.title('Real Brasileiro em Dolares: Lula', weight='bold')
plt.show()
```

```python
### Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Adding the subplots
plt.figure(figsize=(12, 6))
ax1_fhc = plt.subplot(2,5,1)
ax2_lula = plt.subplot(2,5,2)
ax3_dilma = plt.subplot(2,5,3)
ax4_temer = plt.subplot(2,5,4)
ax5_bozo = plt.subplot(2,5,5)

ax6_final = plt.subplot(2,1,2)
axes = [ax1_fhc, ax2_lula, ax3_dilma, ax4_temer, ax5_bozo, ax6_final]

### Changes to all the subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0])
    ax.set_yticklabels(['1.5', '2','2.5', '3', '3.5', '4', '4.5', '5', '5.5', '6'],
                   alpha=0.3)
    ax.grid(alpha=0.5)    
    

### Ax1
ax1_fhc.plot(fhc['Time'], fhc['dollar_rolling_mean'],
        color='#BF5FFF')
ax1_fhc.set_xticklabels(['2000', '', '', '2001', '', '', '2002'],
                   alpha=0.3)
ax1_fhc.text(730300.0, 7.0, 'FHC', fontsize=18, weight='bold',
        color='#BF5FFF')
ax1_fhc.text(730300.0, 6.5, '(2000-2002)', weight='bold',
        alpha=0.3)


### Ax2
ax2_lula.plot(lula['Time'], lula['dollar_rolling_mean'],
        color='#ffa500')
ax2_lula.set_xticklabels(['', '2003', '', '', '', '2007', '', '', '', '2011'],
                   alpha=0.3)
ax2_lula.text(731000.0, 7.0, 'LULA', fontsize=18, weight='bold',
        color='#ffa500')
ax2_lula.text(731000.0, 6.5, '(2003-2010)', weight='bold',
         alpha=0.3)


### Ax3
ax3_dilma.plot(dilma['Time'], dilma['dollar_rolling_mean'],
        color='#00B2EE')
ax3_dilma.set_xticklabels(['', '2012', '', '', '2014', '', '',
                     '2016'],
                   alpha=0.3)
ax3_dilma.text(734500.0, 7.0, 'DILMA', fontsize=18, weight='bold',
        color='#00B2EE')
ax3_dilma.text(734500.0, 6.5, '(2011-2016)', weight='bold',
         alpha=0.3)

## Ax4
ax4_temer.plot(temer['Time'], temer['dollar_rolling_mean'],
        color='#b83333')
ax4_temer.set_xticklabels(['', '2016', '',  '', '2017', '', '', '2018'],
                   alpha=0.3)
ax4_temer.text(736400.0, 7.0, 'TEMER', fontsize=18, weight='bold',
        color='#b83333')
ax4_temer.text(736400.0, 6.5, '(2016-2018)', weight='bold',
         alpha=0.3)

## Ax5
ax5_bozo.plot(bozo['Time'], bozo['dollar_rolling_mean'],
        color='#664710')
ax5_bozo.set_xticklabels(['', '2019', '', '', '2020', '', '', '2021'],
                   alpha=0.3)
ax5_bozo.text(737000.0, 7.0, 'BOLSO', fontsize=18, weight='bold',
        color='#664710')
ax5_bozo.text(737000.0, 6.5, '(2019-2021)', weight='bold',
         alpha=0.3)

### Ax4: Bush-Obama-Trump
ax6_final.plot(fhc['Time'], fhc['dollar_rolling_mean'],
        color='#BF5FFF')
ax6_final.plot(lula['Time'], lula['dollar_rolling_mean'],
        color='#ffa500')
ax6_final.plot(dilma['Time'], dilma['dollar_rolling_mean'],
        color='#00B2EE')
ax6_final.plot(temer['Time'], temer['dollar_rolling_mean'],
        color='#b83333')
ax6_final.plot(bozo['Time'], bozo['dollar_rolling_mean'],
        color='#664710')
ax6_final.grid(alpha=0.5)
ax6_final.set_xticks([])

### Adding a title and a subtitle
ax1_fhc.text(730016.0, 10.0, 'BRL-USD rate under the last 5 Brazilian presidents',
         fontsize=20, weight='bold')
ax1_fhc.text(730016.0, 9.0, '''Taxa de câmbio real-dolar entre os anos 2000 e 2020''', fontsize=16)

### Adding a signature
ax6_final.text(729916.0, 0.65, 'Luiz P. C Alves' + ' '*103 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

plt.show()
```
